```python
def distance(vec1,vec2):
    return (np.sum((vec1-vec2)**2))**0.5
```

```python
def pathlength(path,length):
    dist = 0
    for i in range(1,length):
        dist+=distance(path[i],path[i-1])
    dist+=distance(path[length-1],path[0])
    return dist
```

```python
import random
```

```python
dict ={}
for i in range(10,20):
    dict[i] = []
    shape = np.random.rand(i)
    for j in range(500):
        random.shuffle(shape)
        dict[i].append(pathlength(shape,i))
```

```python
mean_arr = []
std_arr = []
mode_arr = []
```

```python
for i in range(10,20):
    x = sorted(dict[i])
    x_ = []
    x_.append(x[0])
    y = [1]
    for idx in range(1,len(dict[i])):
        if(-x_[len(x_)-1]+x[idx]>0.01):
            x_.append(x[idx])
            y.append(1)
        else:
            y[len(x_)-1]+=1
    
    ##study the std devn and mean
    arr = np.array(x)
    y1 = np.array(y)
    mean_arr.append(arr.mean())
    std_arr.append(arr.std())
    mode_arr.append(x_[np.argmax(y)])
```

```python
plt.plot(list(range(10,20)),mode_arr)
plt.xlabel("no. of cities")
plt.ylabel("mode")
```
